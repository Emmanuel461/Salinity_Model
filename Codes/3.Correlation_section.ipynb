### 2. Run the correlation and linear regression process

```python
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Dividir los datos en características (X) y objetivo (y)
X_corr = datos_corr.drop(columns=['CE'])  # Características
y_corr = datos_corr['CE']  # Objetivo

# Inicializar una lista para almacenar los R^2 y las fórmulas de cada modelo
r2_scores = []
p_values = []
formulas = []

# Iterar sobre cada columna
for column in X_corr.columns:
    # Añadir una constante a las características
    X_with_const = sm.add_constant(X_corr[[column]])

    # Ajustar un modelo lineal a la columna actual
    model = sm.OLS(y_corr, X_with_const)
    result = model.fit()

    # Calcular R^2
    r2 = result.rsquared
    r2_scores.append(r2)

    # Obtener el p-value de la característica
    p_value = result.pvalues[column]
    p_values.append(p_value)

    # Obtener los coeficientes del modelo
    coef = result.params[column]
    intercept = result.params['const'] if 'const' in result.params else 0

    # Crear la fórmula del modelo
    formula = f"SAR = {coef:.2f} * {column} + {intercept:.2f}"
    formulas.append(formula)

    # Imprimir R^2, p-value y fórmula del modelo
    print(f"R^2 para {column}: {r2}")
    print(f"p-value para {column}: {p_value}")
    print(f"Fórmula del modelo para {column}: {formula}")
    print("--------------------------")

# Crear un gráfico de barras para mostrar los R^2
plt.figure(figsize=(10, 6))
plt.barh(X_corr.columns, r2_scores, color='skyblue')
plt.xlabel('R^2 Score')
plt.ylabel('Variables')
plt.title('R^2 Score para cada columna')
plt.show()

# Imprimir p-values de los modelos
print("P-values de los modelos:")
for column, p_value in zip(X_corr.columns, p_values):
    print(f"p-value para {column}: {p_value}")

# Imprimir fórmulas de los modelos
print("Fórmulas de los modelos:")
for formula in formulas:
    print(formula)
```
